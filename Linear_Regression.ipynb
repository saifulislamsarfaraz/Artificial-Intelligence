```python
import matplotlib.pyplot as plt

# Function to calculate mean
def mean(values):
    return sum(values) / len(values)

# Function to calculate slope and intercept
def fit(X, y):
    """
    Train the model by calculating the slope and intercept.
    Formula:
        slope (m) = covariance(X, y) / variance(X)
        intercept (b) = mean(y) - slope * mean(X)
    """
    X_mean = mean(X)
    y_mean = mean(y)

    # Calculate numerator and denominator for the slope
    numerator = sum((X[i] - X_mean) * (y[i] - y_mean) for i in range(len(X)))
    denominator = sum((X[i] - X_mean) ** 2 for i in range(len(X)))

    # Calculate slope and intercept
    slope = numerator / denominator
    intercept = y_mean - slope * X_mean
    return slope, intercept

# Function to predict values
def predict(X, slope, intercept):
    """
    Predict the target values for input features X.
    Formula: y = m*X + b
    """
    return [slope * x + intercept for x in X]

# Function to plot the data and regression line
def plot(X, y, slope, intercept):
    """Plot the data points and the regression line."""
    plt.scatter(X, y, color='blue', label='Data Points')
    predictions = predict(X, slope, intercept)
    plt.plot(X, predictions, color='red', label='Regression Line')
    plt.xlabel('X - Input Feature')
    plt.ylabel('y - Target')
    plt.legend()
    plt.show()

# Sample Dataset (Example)
X = [1, 2, 3, 4, 5]  # Input feature
y = [2, 4, 5, 4, 5]  # Target

# Train (fit) the model
slope, intercept = fit(X, y)

# Print slope and intercept
print(f"Slope (m): {slope}")
print(f"Intercept (b): {intercept}")

# Predict on new data
X_new = [6, 7, 8]
y_pred = predict(X_new, slope, intercept)
print(f"Predictions for {X_new}: {y_pred}")

# Visualize the regression line
plot(X, y, slope, intercept)
```
