```python

import random

def random_chromosome(size):  # Generate random chromosomes
    return [random.randint(1, nq) for _ in range(nq)]

def fitness(chromosome):  # Calculate fitness
    horizontal_collisions = sum([chromosome.count(queen) - 1 for queen in chromosome]) / 2
    diagonal_collisions = 0

    n = len(chromosome)
    left_diagonal = [0] * 2 * n
    right_diagonal = [0] * 2 * n
    for i in range(n):
        left_diagonal[i + chromosome[i] - 1] += 1
        right_diagonal[len(chromosome) - i + chromosome[i] - 2] += 1

    for i in range(2 * n - 1):
        if left_diagonal[i] > 1:
            diagonal_collisions += (left_diagonal[i] - 1)
        if right_diagonal[i] > 1:
            diagonal_collisions += (right_diagonal[i] - 1)

    return int(maxFitness - (horizontal_collisions + diagonal_collisions))

def probability(chromosome, fitness):  # Calculate probability
    return fitness(chromosome) / maxFitness

def random_pick(population, probabilities):  # Select a chromosome randomly
    total = sum(probabilities)
    r = random.uniform(0, total)
    upto = 0
    for chrom, prob in zip(population, probabilities):
        if upto + prob >= r:
            return chrom
        upto += prob
    assert False, "Shouldn't reach here"

def reproduce(x, y):  # Perform two-point crossover
    n = len(x)
    # Select two crossover points
    point1 = random.randint(0, n - 1)
    point2 = random.randint(0, n - 1)
    if point1 > point2:
        point1, point2 = point2, point1
    # Create offspring
    child = x[:point1] + y[point1:point2] + x[point2:]
    return child

def mutate(chromosome):  # Mutate a chromosome
    n = len(chromosome)
    index = random.randint(0, n - 1)
    value = random.randint(1, n)
    chromosome[index] = value
    return chromosome

def genetic_queen(population, fitness):
    mutation_probability = 0.03
    new_population = []
    probabilities = [probability(chrom, fitness) for chrom in population]

    for i in range(len(population)):
        x = random_pick(population, probabilities)  # Parent 1
        y = random_pick(population, probabilities)  # Parent 2
        child = reproduce(x, y)  # Two-point crossover
        if random.random() < mutation_probability:
            child = mutate(child)
        print_chromosome(child)
        new_population.append(child)
        if fitness(child) == maxFitness:
            break
    return new_population

def print_chromosome(chrom):
    print("Chromosome = {}, Fitness = {}".format(chrom, fitness(chrom)))

if __name__ == "__main__":
    nq = int(input("Enter Number of Queens: "))  # Number of queens
    maxFitness = (nq * (nq - 1)) / 2  # Maximum fitness
    population = [random_chromosome(nq) for _ in range(100)]

    generation = 1
    while not maxFitness in [fitness(chrom) for chrom in population]:
        print("=== Generation {} ===".format(generation))
        population = genetic_queen(population, fitness)
        print("")
        print("Maximum Fitness = {}".format(max([fitness(chrom) for chrom in population])))
        generation += 1

    # Display solution
    chrom_out = []
    print("Solved in Generation {}!".format(generation - 1))
    for chrom in population:
        if fitness(chrom) == maxFitness:
            print("")
            print("One of the solutions:")
            chrom_out = chrom
            print_chromosome(chrom)

    # Display chessboard
    board = [["x"] * nq for _ in range(nq)]
    for i in range(nq):
        board[nq - chrom_out[i]][i] = "Q"

    def print_board(board):
        for row in board:
            print(" ".join(row))

    print()
    print_board(board)
```
